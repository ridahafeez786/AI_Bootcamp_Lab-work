## 🔤 Token vs. Character

- A **character** is a single unit: a letter, digit, punctuation mark, emoji, or symbol.
- A **token** is a meaningful chunk of text — usually a word, but can also be a subword, punctuation mark, or symbol — produced by **tokenization** (covered in depth in Part 4).

**Example:**
Text:       "I can't wait!"

Characters: I ,  , c, a, n, ', t,  , w, a, i, t, !     → 13 characters

Tokens:     ["I", "ca", "n't", "wait", "!"]            → 5 tokens (approx., depends on tokenizer)

> ⚠️ **Warning:** Character count and token count are **not** the same thing, and confusing them is a common beginner mistake — especially when working with LLMs, where cost and context limits are measured in *tokens*, not characters or words.

### 3.1 Case Handling — Lowercasing & Case Folding

#### Definition
- **Lowercasing**: Converting all characters to lowercase (`Hello` → `hello`)
- **Case Folding**: A more aggressive, locale-aware version of lowercasing that also handles special Unicode case rules (e.g., German `ß` → `ss`)

#### Why this exists
Machines treat `"Great"`, `"great"`, and `"GREAT"` as three completely different tokens unless normalized. This artificially inflates vocabulary size and splits your signal across multiple tokens.

#### ✅ When to use
- Almost always, for traditional ML pipelines (BoW, TF-IDF)
- Sentiment analysis, topic modeling, search

#### ❌ When NOT to use
- **Named Entity Recognition (NER)** — case is a strong signal (`Apple` the company vs `apple` the fruit)
- **Sarcasm/emphasis detection** — `"I LOVE this"` (sarcastic/emphatic) vs `"i love this"` carries different intensity signal
- Modern **transformer-based models** (BERT, GPT) — these have cased and uncased variants; usually you should **NOT** lowercase before feeding into a cased transformer, since the model learned case as a feature

##### Metacharacters:
- `[]` :	A set of characters , example: "`[a-m]`"
- `\` :	Signals a special sequence (can also be used to escape special characters), example:	`"\d"`
- `.`	: Any character (except newline character), example:	`"he..o"`
- `^`	: Starts with, example: `"^hello"`
- `$`	: Ends with, example:	`"planet\$"`
- `*`	: Zero or more occurrences, example:	`"he.*o"`
- `+`	: One or more occurrences, example:	`"he.+o"`
- `?`	: Zero or one occurrences, example:	`"he.?o"`
- `{}`: Exactly the specified number of occurrences, example:	`"he.{2}o"`
- `|`	: Either or, example:	`"falls|stays"`
- `()`: Capture and group

### 3. Remove HTML Tags

#### Why this exists
Text scraped from websites, emails, or forums often contains HTML markup (`<p>`, `<b>`, `<br>`) that adds visual formatting but no semantic value for most NLP tasks.

#### ✅ When to use
Any time your source is scraped web content, HTML emails, or CMS-exported data.

#### ❌ When NOT to use
If you're building a tool that needs to understand document *structure* (e.g., extracting tables, headers) — in that case, parse the HTML structure first, THEN strip tags from the extracted text.

> 💡 **Best Practice:** Always use a proper HTML parser (`BeautifulSoup`) rather than a naive regex like `re.sub('<.*?>', '', text)`. Regex can fail on malformed HTML, nested tags, or HTML entities (`&amp;`, `&nbsp;`).

As in the review.csv file, we have no HTML tags, so first let's produce them and then include them in csv

````python
import csv
````

````python
input_file = "reviews.csv"
output_file = "reviewsHTML.csv"
````

````python
# Function to convert text to HTML
def text_to_html(text):
    return f"<p>{text}</p>"
````

````python
# Read the input CSV, process it, and write to the output CSV
with open(input_file, mode="r", encoding="utf-8") as infile, \
open(output_file, mode="w", encoding="utf-8", newline="") as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    # Process each row in the input CSV
    for row in reader:
        # Convert each cell in the row to HTML
        html_row = [text_to_html(cell) for cell in row]
        writer.writerow(html_row)

print(f"Converted CSV with HTML saved to {output_file}")
````

### 5. Remove Punctuation

````python
import string
import time
````

````python
def remove_punctuation(text):
    #exclude = string.punctuation  # Predefined punctuation characters
    for char in string.punctuation:
        text = text.replace(char, '')  # Replace punctuation in text
    return text
````

````python
Text = 'Hi! , How are you doing today?'
````

````python
print ('Text with Punctuation is: ',Text)

start = time.time()

remove_punctuation(Text)

print ('Text without Punctuation is: ',remove_punctuation(Text))

ExecTime = time.time()- start

print ('Exec Time for 10 function calls:', ExecTime)

print ('Exec Time repeated for 100,000 function calls:', ExecTime*10000)
````

````python
def remove_punctuationFast(text):
    # Create a translation table that maps all punctuation to None
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
````

````python
print ('Text with Punctuation is: ',Text)
start = time.time()
print ('Text without Punctuation is: ',remove_punctuationFast(Text))
ExecTime = time.time()- start
print ('Exec Time: ', ExecTime)
````

### 6. short_words in chat



````python
short_words_dict = {
    "LOL": "Laughing out loud",
    "BRB": "Be right back",
    "TTYL": "Talk to you later",
    "IMO": "In my opinion",
    "IMHO": "In my humble opinion",
    "IDK": "I don't know",
    "SMH": "Shaking my head",
    "ROFL": "Rolling on the floor laughing",
    "TBH": "To be honest",
    "OMG": "Oh my god",
    "FYI": "For your information",
    "BFF": "Best friends forever",
    "FOMO": "Fear of missing out",
    "YOLO": "You only live once",
    "ICYMI": "In case you missed it",
    "DM": "Direct message",
    "LMAO": "Laughing my ass off",
    "TMI": "Too much information",
    "WTF": "What the heck",
    "NSFW": "Not safe for work",
    "SRS": "Serious"
}
````

````python
# Function to convert short words into complete sentences based on dictionary
def convert_to_complete_sentences(input_text):
    words = input_text.split()  # Split input text into words
    result = []  # Initialize result list to hold converted words
    print(words)

    # Iterate over each word in the input text
    for word in words:
        # Check if the word is in the dictionary
        if word.upper() in short_words_dict:
            result.append(short_words_dict[word.upper()])  # Append the full form
        else:
            result.append(word)  # Append the original word if not found in the dictionary

    # Join the result list into a single string
    return ' '.join(result)
````

````python
# Test the function with an example input
input_text = "LOL. FYI, I am going now. BRB TTYL.DM me if SRS."
converted_text = convert_to_complete_sentences(input_text)
print(converted_text)
````

### 8. Stop words removal

#### 🤔 What Are Stop Words?

**Stop words** are extremely common words that carry little standalone semantic value — words like `"the"`, `"is"`, `"and"`, `"a"`, `"in"`. They appear so frequently across nearly all documents that they add noise rather than signal for many tasks.

> 💡 **Real-world intuition:** If you're trying to figure out what a review is *about* (say, topic classification), the words `"is"`, `"the"`, `"a"` appear in literally every review — positive, negative, about phones, about food. They don't help you distinguish one document from another. But the words `"battery"`, `"terrible"`, `"delicious"` do.

##### Why Remove Them?
1. **Reduce dimensionality** — fewer unique tokens means smaller vocabulary, faster processing, less memory
2. **Reduce noise** — high-frequency, low-information words can dominate frequency-based features (like raw Bag-of-Words counts)
3. **Improve focus** — models trained on content words often generalize better for topic-based tasks

> 📝 **Note:** Stop words are usually language-specific. NLTK, spaCy, and scikit-learn each ship their own stop word lists — and these lists **don't perfectly agree** with each other. There's no single "correct" list.

````python
# Function to remove stop words
def remove_stopwords(text):
    # Tokenize the input text
    words = word_tokenize(text)

    # Get the stop words list for English
    stop_words = set(stopwords.words('english'))

    # Filter out stop words
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Join the filtered words back into a string
    return ' '.join(filtered_words)
````

#### ⚠️ When NOT to Remove Stop Words

This is the most important lesson in this section. Stop word removal is a **default habit** for many beginners — but it actively damages certain tasks.

##### 🔴 Case 1: Sentiment Analysis & Negation

Look what happens to negation:

````python
negation_examples = [
    "this is not good",
    "I am not happy with this product",
    "this was never a good experience",
]

print("The danger of removing stop words for negation-sensitive tasks:\n")
for ex in negation_examples:
    cleaned = remove_stopwords(ex)
    print(f"Original: {ex!r}")
    print(f"Cleaned:  {cleaned!r}")
    print(f"  ⚠️ Notice: '{ex.split()[2] if len(ex.split())>2 else ''}' meaning is now LOST or AMBIGUOUS\n")
````

> 🔴 **This is a critical failure mode.** `"not good"` and `"good"` end up looking nearly identical after naive stop word removal (since `"not"` is in most default stop word lists) — completely flipping the sentiment a model would learn from that phrase.

##### ✅ Fix: Custom Stop Word Lists

The solution isn't to abandon stop word removal — it's to **customize the list** for your task.

##### 🔴 Case 2: Other tasks where stop word removal hurts
| Task | Why stop words matter |
|---|---|
| **Machine Translation** | Function words carry grammatical structure needed for correct translation |
| **Text Generation / LLMs** | Removing stop words produces unnatural, ungrammatical output |
| **POS Tagging / Parsing** | Stop words are essential for grammatical structure analysis |
| **Phrase/Idiom detection** | `"to be or not to be"` — nearly ALL stop words, but highly meaningful |
| **Author style / stylometry** | Function word usage patterns are actually a strong authorship signal |

````python
def build_custom_stopwords(base_list: set, remove_from_list: list = None, add_to_list: list = None) -> set:
    """
    Builds a task-specific stop word list by removing important words
    (like negations) from the base list and/or adding domain-specific noise words.
    """
    custom = set(base_list)
    if remove_from_list:
        custom -= set(remove_from_list)
    if add_to_list:
        custom |= set(add_to_list)
    return custom

# Sentiment-safe stop word list: preserve negation words
negation_words = ["not", "no", "never", "nor", "none", "cannot", "n't", "neither"]
sentiment_safe_stopwords = build_custom_stopwords(nltk_stopwords, remove_from_list=negation_words)

print("Negation words removed from stop list (now preserved during cleaning):", negation_words)
print()

for ex in negation_examples:
    cleaned_naive = remove_stopwords(ex, nltk_stopwords)
    cleaned_safe = remove_stopwords(ex, sentiment_safe_stopwords)
    print(f"Original:       {ex!r}")
    print(f"Naive removal:  {cleaned_naive!r}")
    print(f"Sentiment-safe: {cleaned_safe!r}")
    print()
````

### 9. Tokenization

Tokenization is the process of breaking down text into smaller pieces, called tokens, which
can be words, characters, or subwords

Why is Tokenization Important?
- Computers don't understand raw human language directly. Tokenization converts
unstructured text into a structured format that algorithms can process.


#### Types of Tokenization:
1. Word Tokenization :  Splits text into individual words
2. Character Tokenization : Splits text into individual characters.
3. Subword Tokenization : Splits words into smaller meaningful parts (used in modern
models like BERT or GPT)

Example:
`Original Text: ChatGPT is amazing!`
1. Word Tokenization:
  - `["ChatGPT", "is", "amazing", "!"]`
   Each word and punctuation is treated as a separate token.

2. Character Tokenization:
  - `["C", "h", "a", "t", "G", "P", "T", " ", "i", "s", " ", "a", "m", "a",
"z", "i", "n", "g", "!"]`
  Every character is a token, including spaces and punctuation.

3. Subword Tokenization (e.g., using BPE or WordPiece)
  - `["Chat", "G", "PT", " is", " amazing", "!"]`
  Words are split into smaller chunks, especially useful for rare or compound words

- `(?<=\b)` : ensures the match happens right after a word boundary (start of word).
- `(?=\w+)` : ensures there are more word characters after it.

- `(?<=\w)` : ensures there is at least one word character before it.

- `(?=\b)` : ensures it is right before a word boundary (end of word).

- `(?=\w)` : ensures there’s a word character after it.

##### Regular Expression Lookaround Assertions

Lookarounds allow us to check conditions before or after a pattern without including those characters in the match.

---

###### 1. `(?<=\b)` : Positive Lookbehind (Word Boundary Before)

Ensures that the match occurs immediately after a word boundary (start of a word).

Example: Find words starting after a word boundary

```python
import re

text = "Hello Python World"

result = re.findall(r"(?<=\b)Python", text)

print(result)
```

**Output:**
```
['Python']
```

Explanation:
- `(?<=\b)` checks that `Python` starts at a word boundary.
- The boundary itself is not included in the result.

---

###### 2. `(?=\w+)` : Positive Lookahead (Word Characters After)

Ensures that word characters exist after the current position.

Example: Find "Py" only when followed by word characters.

```python
import re

text = "Python Py 123"

result = re.findall(r"Py(?=\w+)", text)

print(result)
```

**Output:**
```
['Py']
```

Explanation:
- `Py(?=\w+)` matches `Py` only if more word characters follow.
- `Python` matches because `thon` comes after `Py`.
- `Py` alone does not match because no characters follow.

---

###### 3. `(?<=\w)` : Positive Lookbehind (Word Character Before)

Ensures that there is a word character before the match.

Example: Find digits that have a letter before them.

```python
import re

text = "A123 B456 789"

result = re.findall(r"(?<=\w)\d+", text)

print(result)
```

**Output:**
```
['123', '456']
```

Explanation:
- `(?<=\w)` checks the character before the digits.
- `A123` and `B456` match because letters exist before numbers.
- `789` does not match because there is no word character before it.

---

###### 4. `(?=\b)` : Positive Lookahead (Word Boundary After)

Ensures that the match ends at a word boundary.

Example: Find "Python" only when it is a complete word.

```python
import re

text = "Python Python3 Python"

result = re.findall(r"Python(?=\b)", text)

print(result)
```

**Output:**
```
['Python', 'Python']
```

Explanation:
- First `Python` → followed by a space, so boundary exists ✅
- `Python3` → followed by `3`, so no boundary ❌
- Last `Python` → end of string, so boundary exists ✅

---

###### 5. `(?=\w)` : Positive Lookahead (Word Character After)

Ensures that a word character exists immediately after the match.

Example: Find "Py" only when followed by a word character.

```python
import re

text = "Python Py Py@"

result = re.findall(r"Py(?=\w)", text)

print(result)
```

**Output:**
```
['Py']
```

Explanation:
- `Python` → `Py` is followed by `t` (word character) ✅
- `Py` → nothing follows ❌
- `Py@` → `@` is not a word character ❌

---

###### Summary Table

| Lookaround | Meaning | Example |
|------------|---------|---------|
| `(?<=\b)` | Match starts after word boundary | `(?<=\b)Python` |
| `(?=\w+)` | Word characters exist after match | `Py(?=\w+)` |
| `(?<=\w)` | Word character exists before match | `(?<=\w)\d+` |
| `(?=\b)` | Match ends at word boundary | `Python(?=\b)` |
| `(?=\w)` | Word character exists after match | `Py(?=\w)` |

##### Tokenize a sentence while identifying prefixes, suffixes, and infixes using Regular Expressions.

It tries to split words based on common word parts like:

* **Prefix**: `un`, `pre`, `re`, `mis`, `anti`
* **Suffix**: `ly`, `ing`, `ed`, `s`, `es`
* **Infix**: `-` inside hyphenated words

````python
import re
def tokenize_with_affixes(text):
    # Define patterns for prefix, suffix, and infix
    prefix_pattern = r'(?<=\b)(un|pre|re|mis|anti)(?=\w+)' # Example prefixes
    suffix_pattern = r'(?<=\w)(ly|ing|ed|s|es)(?=\b)'      # Example suffixes
    infix_pattern = r'(?<=\w)-(?=\w)'                      # Example infixes
                                                           # (like hyphenated words)
    # Split based on the patterns
    tokens = re.split(f"{prefix_pattern}|{suffix_pattern}|{infix_pattern}", text)
    # Filter out empty tokens and return
    return [token for token in tokens if token]
````

````python
# Example 1
text = "Hello!, I am going to Lahore tomtorrow :-)."
tokens = tokenize_with_affixes(text)
print(tokens)

text = "Hello, I am going to Lahore tomorrow :-)."
tokens = tokenize_with_affixes(text)
print(tokens)
````

### 10. Stemming
Reduces words to their root form by chopping off suffixes.

#### 🤔 The Need for Stemming

Words like `"run"`, `"running"`, `"runs"`, and `"ran"` all refer to the same core concept, but a machine treats each as a **completely different token** unless we normalize them.

> 💡 **Real-world intuition:** Imagine searching a document store for "connect" — wouldn't you also want results containing "connecting", "connected", "connection"? Stemming makes this kind of matching possible by reducing all these variants down to a shared root form.

**Stemming** is a rule-based, often crude process of chopping word endings to arrive at a "stem" — which may not even be a real word.

running   -> run

runs      -> run

ran       -> ran (irregular forms are NOT normalized by stemming!)

connection -> connect

connected  -> connect

connecting -> connect

> ⚠️ **Warning:** Notice `"ran"` stayed as `"ran"` — stemming uses simple suffix-stripping rules, NOT dictionary lookups or grammar understanding. It cannot handle irregular forms. This is a key limitation we'll revisit when comparing to lemmatization.

#### 🔬 Stemming Algorithms

There are three classic stemming algorithms, each with different levels of aggressiveness.

| Algorithm | Approach | Aggressiveness |
|---|---|---|
| **Porter** | Oldest (1980), widely used, moderate rule set | Moderate |
| **Snowball** (Porter2) | Improved/refined version of Porter, more consistent | Moderate (slightly more precise than Porter) |
| **Lancaster** | Very aggressive, iterative rule application | Very aggressive |

#### 10.1. PorterStemmer
   - Best for general English text processing tasks like information retrieval, search engines,
or basic NLP pipelines.
  - Good balance between performance and accuracy.
  - May not always produce real words.

````python
from nltk.stem.porter import PorterStemmer
````

````python
stemmer = PorterStemmer()

words_list = ["running", "flies", "easily", "flying", "played", "happily"]

stem = [stemmer.stem(word) for word in words_list]

print(stem)
````

#### ✅ When to Use Stemming

- **Information retrieval / search engines** — matching "connect" with "connecting", "connected" improves recall
- Large-scale text indexing where speed matters more than linguistic precision
- Preprocessing for simple BoW models where slight imprecision is an acceptable trade-off for speed

#### ❌ When NOT to Use Stemming

- **Any task requiring human-readable output** (summarization, generation, chatbots) — stems like `"studi"` or `"nation"` look broken to end users

- When you have computational budget for **lemmatization** instead (Part 7) — lemmatization is almost always linguistically superior when speed isn't the bottleneck

### 11. Lemmatization

#### 🤔 The Need for Lemmatization

Like stemming, **lemmatization** reduces words to a base form so that `"running"`, `"runs"`, and `"ran"` are all recognized as related to `"run"`. But *how* it gets there is fundamentally different.

> 💡 **Real-world intuition:** Stemming is like using a paper cutter — fast, crude, chops off endings based on simple rules, and sometimes produces something that isn't even a real word (`"studies"` → `"studi"`). Lemmatization is like asking a dictionary editor — slower, but it always returns a real, valid dictionary word (the **lemma**), using actual grammar and vocabulary knowledge.

| | Stemming | Lemmatization |
|---|---|---|
| **Method** | Rule-based suffix stripping | Dictionary + grammar-based lookup |
| **Output** | May not be a real word (`"studi"`) | Always a real word (`"study"`) |
| **Speed** | Fast | Slower (needs POS/context) |
| **Accuracy** | Lower — many mis-conflations | Higher — linguistically correct |
| **Needs context?** | No | Often yes (POS tag) |

#### 🏷️ Why POS Tags Matter for Lemmatization

The SAME word can have a different lemma depending on its **part of speech** (its grammatical role in the sentence).

**Example: "meeting"**
- As a **noun**: `"I attended the meeting"` → lemma = `"meeting"` (a scheduled event)
- As a **verb**: `"I am meeting him tomorrow"` → lemma = `"meet"` (the base verb)

If you lemmatize without telling the lemmatizer the POS tag, it defaults to treating everything as a **noun** — which silently produces wrong results for verbs, adjectives, and adverbs.

### 🆚 Stemming vs. Lemmatization — Direct Side-by-Side Comparison

````python
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
````

````python
from nltk.stem import PorterStemmer
from nltk import pos_tag
from nltk.corpus import wordnet

# Quick stemmer instance for comparison purposes (full coverage in Part 6)
stemmer = PorterStemmer()

comparison_words = ["studies", "studying", "better", "geese", "running",
                     "was", "meeting", "universities", "wolves", "caring"]

def get_wordnet_pos(nltk_pos_tag: str):
    """Maps NLTK's POS tag format to WordNet's simplified POS format."""
    if nltk_pos_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_pos_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_pos_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_pos_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # default fallback


print(f"{'Word':<15}{'Stemmed':<15}{'Lemmatized (POS-aware)':<25}")
print("-" * 55)
for w in comparison_words:
    stemmed = stemmer.stem(w)
    tagged = pos_tag([w])
    lemma = lemmatizer.lemmatize(w, get_wordnet_pos(tagged[0][1]))
    print(f"{w:<15}{stemmed:<15}{lemma:<25}")
````

> 📝 **Note the key differences in the table above:**
> - `"studies"` → stemmed to `"studi"` (not a real word!) vs. lemmatized correctly to `"study"`
> - `"better"` → stemming barely touches it (`"better"`) vs. lemmatization correctly maps to `"good"` (its true root)
> - `"geese"` → stemming can't handle irregular plurals (`"gees"` or unchanged) vs. lemmatization correctly returns `"goose"`
> - `"wolves"` → similar irregular plural handling difference

#### 📊 When to Use Which

| Scenario | Recommended |
|---|---|
| Search engines (speed matters, some noise tolerable) | **Stemming** |
| Sentiment analysis, text classification | **Lemmatization** (more accurate signal) |
| Large-scale, real-time systems with huge corpora | **Stemming** (much faster) |
| Chatbots, question-answering, any human-facing accuracy | **Lemmatization** |
| Quick prototyping / exploratory analysis | **Stemming** (simpler, no POS tagging needed) |
| Any task where output readability matters | **Lemmatization** (always real words) |

> ⚠️ **Warning:** Lemmatization is **significantly slower** than stemming because it requires POS tagging and dictionary lookups. For very large corpora (millions of documents) with tight latency requirements, this performance cost is a real production consideration — not just an academic detail.
